## 3. Cleaning

In this section, the previously mentioned problems would be fixed (except for the name problems), but any ways the name columns would not be important in the analysis and visualisation steps. The order of the cleaning steps would not be the same as the mentioned assessment notes, as some of the steps should be done first, and some after merging the databases, and some after dividing the master dataframe into two dataframes.

#### 1- Solving the issue of double ratings (Quality)

Using regex, the tweets with more than one rating were found and the rating was replaced with NaN values in a new column for corrected numerators.

#### 2- Solving the issue of decimal numerators (Quality)

All tweets with decimal numerators were found using regex, and then the numerator values were corrected in the corrected numerator column. The old numerator column was replaced by the new one. 
#### (note the difference between the old and corrected values below)

```python
twitter_archive_clean
```

#### 3- Merging the numerator/denominator into one column (Quality)
Now with the rating numerator corrected, there is no need for two columns for numerator/denominator. merging them using concatination with '/' would make it string, which is inconvenient because it is needed in calculation and visualization after. so I divided the numerator/denominator and multiplied by 10, and dropped the numerator/denominator columns.

#### 4- Merging the 3 dataframes (Tidiness)

Then, the 3 dfs needed to be merged for further cleaning, so that some entries would be deleted from the 3 dfs. a copy was made.

#### 5- Removing retweets/replies (Quality)
Since we only need original tweets, retweets/replies were removed using the 'in_reply_to_status_id' and 'retweeted_status_id' columns.

#### 6- Removing htttp code from the source column using regex (Quality)
The http code surronding the tweet source in 'source' column is removed using regex.

#### 7- removing columns we do not need (Quality)

Columns 'in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id','retweeted_status_user_id', 'retweeted_status_timestamp' were removed.

#### 8- Dropping non-dog entries (according to the first prediction) (Quality)
In the predicitions df, some of the photos are not of dogs according to the algorithm (1st prediciton). those entries need to be removed using the False values of 'p1_dog'.

#### 9- Dropping p1_dog p2 p2_conf p2_dog p3 p3_conf p3_dog columns (Quality)
Since we only used the 1st prediction, the above mentioned columns are redundant. p1 and p1_conf were renamed appropriately.

#### 10- Correcting Invalid Datatypes (Quality)
Some datatypes are corrected (tweet_id to object and timestamp to datetime).

#### 11- Dividing the master dataframe into 2 frames (one for the tweets, and the other for the dogs) [Tidiness]
A master dataframe is needed in plots, but dividing the data into two separate dfs for dogs and tweets will help the data to be clearer and meaningful, and they still can be joined by the key column (tweet_id)

#### 12- Merging the dog stages columns into one (Tidiness)
Having 4 columns for dog stages is redundant and actually confusing. Having 1 column makes the data tidy. dogs with unidentifyed stage are marked with a NaN value, and for those with multiple stages, all stages are mentioned. The original 4 columns are dropped.

#### 13- Making the breed_prediction column lowercase (Quality)
breed_prediction column had inconsistent capitalization, so I made all of them in lowercase.
